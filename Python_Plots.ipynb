Linear Graph

```python
import numpy as np
#define x & y axsis
x = np.arange(1, 10) #starting value & number of intervals
y = 1.25*x + 1

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y)

ax.set_title("This is a normal plot")
ax.set_xlabel("X values")
ax.set_ylabel("Y values")

plt.show()
```

Different Liner & non-linear models

```python
constant_list = np.ones(len(x))
squared_list = x ** 2
cubed_list = x ** 3

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, constant_list)

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, squared_list)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, cubed_list)

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(x, constant_list, label="Constant")

ax2 = fig.add_subplot(1, 1, 1)
ax2.plot(x, y, label="Linear")

ax3 = fig.add_subplot(1, 1, 1)
ax3.plot(x, squared_list, label="Squared")

ax4 = fig.add_subplot(1, 1, 1)
ax4.plot(x, cubed_list, label="Cubed")

plt.legend()
plt.show()
```

Histogram - Normal Distribution

```python
rand_vals = np.random.normal(5, 2, 1000)

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
ax.hist(rand_vals)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x, constant_list)
axes[0, 1].plot(x, y)
axes[1, 1].plot(x, squared_list)
axes[1, 0].plot(x, cubed_list)
axes[1, 1].set_ylim(-30,200)
plt.show()
```

```python
X = np.linspace(0, 10, 1000)

fig = plt.figure(figsize=(12, 8))

for i in range(6):
  ax = fig.add_subplot(2, 3, i + 1)
  fig.axes[i].plot(X, np.sin(i*X))
  ax.set_title(f"The plot number: {i+1}")

fig.subplots_adjust(wspace=0.2, hspace=0.2)
```

```python
ig, ax = plt.subplots(2, 3, sharex=True, sharey=True)

for i in range(2):
  for j in range(3):
    ax[i, j].plot(X, np.sin(i*X))
```

```python
fig, ax = plt.subplots()
ax.plot(X, X**(1/2))
# ax.get_xticks()
ax.set_xticklabels(labels=ax.get_xticks(), rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# set seed for reproducing
np.random.seed(42)
n = 5000
mean_mu1 = 60
sd_sigma1 = 15
data1 = np.random.normal(mean_mu1, sd_sigma1, n)
mean_mu2 = 80
sd_sigma2 = 15
data2 = np.random.normal(mean_mu2, sd_sigma2, n)

plt.figure(figsize=(8,6))
plt.hist(data1, bins=100, alpha=0.5, label="data1")
plt.hist(data2, bins=100, alpha=0.5, label="data2")
plt.hist(data1+data2, bins=100, alpha=0.5, label="data3")
plt.xlabel("Data", size=14)
plt.ylabel("Count", size=14)
plt.title("Multiple Histograms with Matplotlib")
plt.legend(loc='upper right')
plt.savefig("overlapping_histograms_with_matplotlib_Python_2.png")
```
